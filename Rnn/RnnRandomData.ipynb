```python
import torch
from Rnn_Layers import *
from Rnn_model import *
# Données
vocab_size = 7
seq_len = 5
x = torch.randn(10000, seq_len, vocab_size)
w_true = torch.randn(vocab_size, vocab_size)  # vrai mapping linéaire
y_scores = torch.matmul(x, w_true)  # projection linéaire
y = y_scores.argmax(dim=-1) 
one_hot_y = torch.eye(vocab_size)[y]
```

```python
x_train, y_train, x_test, y_test = split_data(x, one_hot_y, 0.7)

# Modèle
rnn_model = Rnn_model("adam", "Crossentropy")
rnn_model.add_layers([
    Gru(vocab_size,64 ),
    Layer(64,7,None)  ])

losses = rnn_model.train_rnn_sgd(
    epochs=10,
    x_train=x_train,
    y_train=y_train,
    x_val=x_test,
    y_val=y_test,
    batch_size=32,
    learning_rate=0.1,
    accuracy=True
)
```
